let us bin the categories to overcome the possible problems due to rare ones

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
house_type = df[['Konut Şekli', 'rent']].groupby('Konut Şekli').mean()
house_type['count'] = df[['Konut Şekli', 'rent']].groupby('Konut Şekli').count()['rent']
house_type
```

```python
plt.bar(house_type.index, house_type['rent'])
plt.xticks(house_type.index, rotation=90)
plt.show()
```

```python
labels = list(zip(list(house_type.index), list(house_type['count'])))
dendrogram = sch.dendrogram(sch.linkage(np.array(house_type['rent']).reshape(-1,1), method='ward'), labels=labels, leaf_rotation=90)
```

```python
houses = house_type.loc[['Kooperatif', 'Prefabrik', 'Daire', 'Müstakil Ev', 'Köy Evi', 'Çiftlik Evi']]

plt.subplot(1, 2, 1)
labels = list(zip(list(houses.index), list(houses['count'])))
dendrogram = sch.dendrogram(sch.linkage(np.array(houses['rent']).reshape(-1,1), method='ward'), labels=labels, leaf_rotation=90)

plt.subplot(1, 2, 2)
plt.bar(houses.index, houses['rent'])
plt.xticks(houses.index, rotation=90)
plt.show()
```

```python
houses = house_type.loc[['Köşk', 'Bina', 'Residence', 'Yazlık', 'Villa', 'Yalı Dairesi']]

plt.subplot(1, 2, 1)
labels = list(zip(list(houses.index), list(houses['count'])))
dendrogram = sch.dendrogram(sch.linkage(np.array(houses['rent']).reshape(-1,1), method='ward'), labels=labels, leaf_rotation=90)

plt.subplot(1, 2, 2)
plt.bar(houses.index, houses['rent'])
plt.xticks(houses.index, rotation=90)
plt.show()
```

```python
# binning based both on domain knowledge and the above figures

categories = {0: ['Müstakil Ev', 'Köy Evi', 'Çiftlik Evi'],
              1: ['Daire'],
              2: ['Kooperatif', 'Prefabrik'],
              3: ['Villa', 'Yalı Dairesi', 'Yazlık'],
              4: ['Residence', 'Bina'],
              5: ['Köşk']}

categories_map = {}
for k, v in categories.items():
    for cat in v:
        categories_map[cat] = k
# categories_map
```

```python
house_type['binned_category'] = list(house_type.reset_index()['Konut Şekli'].map(categories_map))
house_type
```

```python
house_type_binned = house_type[['count', 'binned_category']].groupby('binned_category').sum()
house_type_binned['categories_involved'] = list(house_type_binned.reset_index()['binned_category'].map(categories))
house_type_binned
```

we drop categories 2 and 5 as they are rare and it's not reasonable to bin them with other categories

```python
df['binned_konut_sekli'] = df['Konut Şekli'].map(categories_map)
df = df[~df['binned_konut_sekli'].isin([2, 5])]
df[['Konut Şekli', 'binned_konut_sekli']]
```

# Initial Model Fitting

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
df_model.fillna(-10000, inplace=True)
X, y = df_model.drop('rent', axis=1), df_model['rent']
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

```python
rf = RandomForestRegressor().fit(X_train, y_train)
```

```python
rf.score(X_train, y_train), rf.score(X_test, y_test)
```

```python
gb = GradientBoostingRegressor().fit(X_train, y_train)
```

```python
gb.score(X_train, y_train), gb.score(X_test, y_test)
```
